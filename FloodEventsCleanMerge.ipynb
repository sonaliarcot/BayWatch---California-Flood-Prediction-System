```python
import pandas as pd
from geopy.geocoders import Nominatim

# create a geocoder object
geolocator = Nominatim(user_agent="sk-application")

# define a function to get latitude and longitude for a given location string
def get_lat_lon(location_str):
    location = geolocator.geocode(location_str)
    if location is None:
        return None, None
    else:
        return location.latitude, location.longitude

# iterate over the rows of the dataframe and fill in missing latitude and longitude values
for i, row in d1.iterrows():
    if pd.isna(row['begin_lat']) or pd.isna(row['begin_lon']):
        location_str = row['begin_location'] + ', ' + row['cz_name_str'] + ', USA'
        lat, lon = get_lat_lon(location_str)
        d1.loc[i, 'begin_lat'] = lat
        d1.loc[i, 'begin_lon'] = lon

print(d1)
```
